```python
#Average rating per movie
rating_avg = rating[['movie id','rating']].groupby('movie id')['rating'].mean().reset_index()
rating_avg.columns = ['movie id','rating_avg']
rating_copy=pd.merge(rating_counts, rating_avg, on='movie id')
rating_copy

#ating_avg
```

```python
rating_count100 = rating_copy[rating_copy['rating_count']>100]
rating_count100
```
